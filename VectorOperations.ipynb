Plotting a vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of two vectors

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
sum = vector_1+vector_2
```

```python
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1, color = 'b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1, color = 'r')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1, color = 'g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of two vectors

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
difference = vector_1-vector_2
print(difference)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1, color = 'b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1, color = 'r')
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1, color = 'g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Cross Product of two vectors

```python
a = np.array([2,3])
b = np.array([4,4])
a_cross_b = np.cross(a,b)
```

```python
print(a_cross_b)
```

```python
c = np.array([2,3,4])
d = np.array([5,10,20])
c_cross_d = np.cross(c,d)
print(c_cross_d)
```

Projection of a vector on v vector

```python
a = np.array([2,5])
v = np.array([3,-4])

#magnitude pf v vector
magnitude_of_v = np.sqrt(v[0]**2+v[1]**2)

#projection of a vector on v vector
projection_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v

print(projection_of_a_on_v)
```

```python
a = np.array([23,45,62])
v = np.array([45,82,67])

#magnitude pf v vector
magnitude_of_v = np.sqrt(v[0]**2+v[1]**2+v[2]**2)

#projection of a vector on v vector
projection_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v

print(projection_of_a_on_v)
```
