Matplotlib

```python
# import matplotlib

import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plot

import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
y = np.sin(x)
```

```python
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#Plotting a data
```

```python
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title x-axis and y-axis labels
plt.plot(x , y )
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
x = np.linspace(-5 ,5 ,50)
plt.plot(x,np.sin(x) , 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

Bar plot


```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['english','french','spanish','latin','german']
people = [100 , 50 , 150 , 40 , 70]
ax.bar(languages , people)
plt.xlabel('languages')
plt.ylabel('no of people')
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['english','hindi','sanskrit','french','mars']
people = [100,50,45,65,23]
ax.pie(people, labels=languages , autopct='%1.1f%%')
plt.show()
```

```python
#Scatter Plot

x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2= plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
