Handling Imbalanced Dataset

A dataset with an unequal class distribution

```python
import numpy as np
import pandas as pd
```

```python
#laoding dataset to pandas dataframe

credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

0 -> legit transaction
1 -> fraudulant transaction

```python
credit_card_data.tail()
```

```python
# determine the distribution of the two classes

credit_card_data['Class'].value_counts()
```

this is highly imbalanced dataset

```python
#Seperating the legit and fraudulant transaction
legit = credit_card_data[credit_card_data.Class == 0] #it will take all the rows which has class value as row
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

UnderSampling

Building a sample dataset containing the similar distribution of legit and fraudulant transaction


```python
#Number of fraudulant transaction -> 232

legit_sample = legit.sample(n=232) #it will get 232 random values and it will store it in legit sample
```

```python
print(legit_sample.shape)
```

```python
#Concatinate the two Dataframe

new_dataset = pd.concat([legit_sample, fraud], axis=0)
```
